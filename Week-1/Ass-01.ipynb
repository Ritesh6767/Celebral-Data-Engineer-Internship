## Handling missing values

### drop columns which have null more than 30%

```python
df.drop(columns=['what_customers_said'], inplace=True)
df.drop(columns=["videos"], inplace=True)
df.drop(columns=["variations"], inplace=True)
```

```python

df.drop(columns=["seller_information"], inplace=True)
df.drop(columns=["seller_name"], inplace=True)
```

### Filling the columns

```python
df["discount"].fillna(df["discount"].median(), inplace=True)
```

## Data formatting

```python
# Normalize final_price values and convert to numeric
final_price_clean = (
    df['final_price']
      .astype(str)
      .str.replace('₹', '', regex=False)
      .str.replace('"', '', regex=False)
      .str.replace(',', '', regex=False)
)

df['final_price'] = pd.to_numeric(final_price_clean, errors='coerce')
# Use initial_price as fallback when final_price could not be parsed
df.loc[df['final_price'].isna(), 'final_price'] = df['initial_price']
```

```python
# Rename the product title column to a business-friendly label
# This updates the DataFrame column label.
df.rename(columns={"title": "Brand_name"}, inplace=True)
```

```python
import ast
import pandas as pd
def clean_best_offer(value):
    if pd.isna(value):
        return "No best offer available"
    s = str(value).strip()
    
    try:
        parsed = ast.literal_eval(s)
        texts = []
        if isinstance(parsed, list):
            for item in parsed:
                if isinstance(item, dict):
                    
                    for key in ('text','offer','best_offer','description','value'):
                        v = item.get(key)
                        if v:
                            texts.append(str(v).strip())
                            break
                    else:
                        texts.extend([str(v).strip() for v in item.values() if v])
                else:
                    texts.append(str(item).strip())
        elif isinstance(parsed, dict):
            for key in ('text','offer','best_offer','description','value'):
                if key in parsed and parsed[key]:
                    texts.append(str(parsed[key]).strip())
                    break
            else:
                texts.extend([str(v).strip() for v in parsed.values() if v])
        cleaned = ', '.join([t for t in texts if t and t not in ('', '\"\"')])
        if cleaned:
            return cleaned
    except Exception:
        pass
    
    cleaned = s.replace('[','').replace(']','').replace('{','').replace('}','').replace('\\\"','').replace("'",'').strip()
    if cleaned in ('', '[]', '[ ]', '{ }', '{}'):
        return "No best offer available"
    return cleaned


df['best_offer'] = df['best_offer'].apply(clean_best_offer)
```

```python
import ast
def extract_sizes(value):
    if pd.isna(value):
        return value
    try:
        parsed = ast.literal_eval(value)
        sizes = []
        for item in parsed:
            if isinstance(item, dict):
                size = item.get('size')
                if size and size not in sizes:
                    sizes.append(size)
        return ', '.join(sizes) if sizes else value
    except Exception:
        return value

df['sizes'] = df['sizes'].apply(extract_sizes)
```

## Basic operation

```python
# Filter rows with valid numeric final_price values and select relevant columns
selected_columns = [
    'product_id',
    'Brand_name',
    'category',
    'initial_price',
    'final_price',
    'discount',
    'sizes',
    'best_offer',
    'rating',
    'ratings_count'
]

clean_df = df.loc[
    df['final_price'].notna() & (df['final_price'] > 0),
    selected_columns
].copy()

clean_df.rename(
    columns={'ratings_count': 'rating_count'},
    inplace=True
)

clean_df.head()
```

## Duplicate Data

```python
# Remove duplicate records from the cleaned dataset
print('Duplicates before:', clean_df.duplicated().sum())

clean_df.drop_duplicates(inplace=True)
print('Duplicates after:', clean_df.duplicated().sum())
```

## Saving Clean Dataset

```python
clean_df.to_csv('cleaned_dataset.csv', index=False)
print("Saved cleaned dataset with shape:", df.shape)
```
